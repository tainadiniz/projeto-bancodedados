```python
import pandas as pd
from sqlalchemy import create_engine  
import mysql.connector

df = pd.read_csv("SuperstoreDataset.csv")
df.info()
```

```python
# Criando um dicionário para alterar o nome ds colunas em portugues
mapeamento_colunas = {
    'Ship Mode': 'Modo de Envio',
    'Customer ID': 'ID do Cliente',
    'Customer Name': 'Nome do Cliente',
    'Segment': 'Segmento',
    'Country': 'País',
    'City': 'Cidade',
    'State': 'Estado',
    'Postal Code': 'Código Postal',
    'Region': 'Região',
    'Product ID': 'ID do Produto',
    'Category': 'Categoria',
    'Sub-Category': 'Subcategoria',
    'Product Name': 'Nome do Produto',
    'Sales': 'Vendas',
    'Quantity': 'Quantidade',
    'Discount': 'Desconto',
    'Profit': 'Lucro',
    'Unnamed: 17': 'Coluna Desconhecida'
}

df.rename(columns=mapeamento_colunas, inplace=True)


df
```

```python
#Apagando a coluna poluída
df.drop(columns=['Coluna Desconhecida'], inplace=True)
```

```python
df.info()
```

```python
# Criando tabela cliente
cliente = df[['ID do Cliente', 'Nome do Cliente', 'Segmento', 'País', 'Cidade', 'Estado', 'Código Postal', 'Região']]
cliente

```

```python
produtos= df[['ID do Produto','Categoria','Subcategoria','Nome do Produto']]
produtos
```

```python
pedido = df[['Modo de Envio','Vendas','Quantidade','Desconto','Lucro']]
pedido
```

```python
#Fazendo conexão python com mysql 


host = 'localhost'
user = 'root'
password = 'admin'
database = 'ecommerce'

engine = create_engine(f'mysql+mysqlconnector://{user}:{password}@{host}/{database}')

```

```python
# Inserir dados nas tabelas
cliente.to_sql('Clientes', con=engine, if_exists='append', index=False)
produtos.to_sql('Produtos', con=engine, if_exists='append', index=False)
pedido.to_sql('Pedidos', con=engine, if_exists='append', index=False)
```

```python
#Fechando a conexão
engine.dispose() 
```
